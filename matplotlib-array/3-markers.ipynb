```python
import matplotlib.pyplot as plt
```

```python
y = [3,8,1,10]

plt.plot(y, marker = 'o')
plt.show()
```

```python
plt.plot(y, marker = '*')
plt.show()
```

```python
plt.plot(y, 'o:r')
plt.show()
```

```python
plt.plot(y, marker = 'o', ms = 20, mec = 'red')
plt.show()
```

```python
plt.plot(y, marker = 'o', ms = 20, mfc = 'r')
```

```python
plt.plot(y, marker = 'o', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')
```
